#### Projeto 01 - Automatização de uma tarefa
##### Neste exemplo foi automatizado o calculo de dados de uma planilha e envio de e-mail

1- Baixar o arquivo excel que está em uma pasta drive (https://drive.google.com/drive/folders/1EtUOFnU8-G6995LGxYxya_I6PlXV4Sml);<br>
2- Trabalhar o resultado da venda na planilha (soma do "Faturamento" e "Quantidade de Produtos");<br>
3- Enviar por e-mail.

```python
import pyautogui
import pyperclip
import time
import pandas as pd

pyautogui.PAUSE = 2

drive = "https://drive.google.com/drive/folders/1EtUOFnU8-G6995LGxYxya_I6PlXV4Sml"
gmail = "https://mail.google.com/mail/u/0/#inbox"
email = "email@email.com"  # inserir e-mail destino que será enviado
assunto = "Relatório de Vendas"

# Passo 1- Entrar no drive e baixar a planilha
# pyautogui.alert("O programa irá iniciar, aperte OK e aguarde o fim do processo.")
pyautogui.hotkey("ctrl", "t")
pyperclip.copy(drive)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

# Passo 2- Localizar a planilha para baixar
# Foram utilizados coordenadas na tela para realizar os cliques no drive para baixar o arquivo
pyautogui.click(x=496, y=433)
pyautogui.click(x=1665, y=237)
pyautogui.click(x=1398, y=740)
time.sleep(3)
pyautogui.click(x=1115, y=631)
```

```python
# Passo 2- Armazenar e alcular os resultados necessários da planilha
# Precisa indicar o caminho onde foi baixado a planilha
caminho = r"C:\Users\[...]\Vendas - Dez.xlsx"
planilha = pd.read_excel(caminho)
display(planilha)

faturamento = planilha["Valor Final"].sum()
quantidade = planilha["Quantidade"].sum()
```

```python
# Passo 3- Abrir e enviar o e-mail com o resultado
texto = f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abraço.
Joel Duarte
"""

# Foram utilizados atalhos para a maioria das ações, precisando clicar apenas no botão de escrever e-mail
pyautogui.hotkey("ctrl", "t")
pyperclip.copy(gmail)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(5)
pyautogui.click(x=104, y=263)
pyautogui.write(email)
pyautogui.press("tab")
pyautogui.press("tab")
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("ctrl", "enter")
```
